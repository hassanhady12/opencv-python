```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    img = cv2.imread('img/bricks.jpg').astype(np.float32) / 255
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
#load_img()
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax  = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
#isplay_img(img)
```

```python
i = load_img()
display_img(i)
```

```python
gamma = 2
```

```python
gamma
```

```python
result = np.power(i,gamma)
```

```python
display_img(result)
```
