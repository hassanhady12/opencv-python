```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('img/dog_backpack.png')
img1.shape

img1 = np.asarray(img1)
img1
```

```python
plt.imshow(img1)
```

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

plt.imshow(img1)
```

```python
img2 = cv2.imread('img/watermark_no_copy.png')
img2.shape
```

```python
plt.imshow(img2)
```

```python
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

plt.imshow(img2)
```

```python
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blend = cv2.addWeighted(src1=img1,alpha=0.4,src2=img2,beta=0.6,gamma=0)
```

```python
plt.imshow(blend)
```
